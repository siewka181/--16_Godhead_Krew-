```python

import jax
import jax.numpy as jnp
import pickle
import os
import time

# === SYSTEM KRÓLÓW I HIERARCHII v1 ===
HIERARCHY_NAMES = ["Lud", "Szlachta", "Książę", "KRÓL"]

def load_swarm_with_hierarchy():
    if os.path.exists(SWARM_STATE_PATH):
        with open(SWARM_STATE_PATH, 'rb') as f:
            data = pickle.load(f)
        traits = data["traits"]
        factions = data.get("factions", jnp.zeros(N_AGENTS, dtype=int))
        legends = data.get("legends", jnp.full(N_AGENTS, -1, dtype=int))
        hierarchy = data.get("hierarchy", jnp.zeros(N_AGENTS, dtype=int))  # 0=Lud, 1=Szlachta, 2=Książę, 3=Król
        print(f">>> RÓJ WCZYTANY Z HIERARCHIĄ – Królów: {jnp.sum(hierarchy == 3)}")
    else:
        traits = swarm["traits"] if "swarm" in globals() else jnp.zeros((N_AGENTS, 10))
        factions = jnp.zeros(N_AGENTS, dtype=int)
        legends = jnp.full(N_AGENTS, -1, dtype=int)
        hierarchy = jnp.zeros(N_AGENTS, dtype=int)
    return {"traits": traits, "factions": factions, "legends": legends, "hierarchy": hierarchy}

swarm = load_swarm_with_hierarchy()
traits = swarm["traits"]
factions = swarm["factions"]
legends = swarm["legends"]
hierarchy = swarm["hierarchy"]

def assign_hierarchy(traits, factions, cycle):
    """Automatyczne nadawanie rang w każdej frakcji"""
    global hierarchy
    num_factions = 4
    hierarchy = hierarchy.at[:].set(0)  # reset na Lud

    for f in range(num_factions):
        mask = (factions == f)
        if not jnp.any(mask):
            continue
        faction_traits = traits[mask]
        faction_idx = jnp.where(mask)[0]

        # Król = najlepszy w frakcji (wysoki INT + Przebiegłość + Percepcja)
        king_score = faction_traits[:, 2] * 0.5 + faction_traits[:, 6] * 0.3 + faction_traits[:, 9] * 0.2
        king_local = jnp.argmax(king_score)
        king_global = faction_idx[king_local]
        hierarchy = hierarchy.at[king_global].set(3)  # KRÓL

        # Szlachta = top 8% w frakcji
        top_szlachta = jnp.argsort(king_score)[-int(len(faction_idx)*0.08):]
        hierarchy = hierarchy.at[faction_idx[top_szlachta]].set(1)

        # Książęta = top 2% pod królem
        top_ksiaze = jnp.argsort(king_score)[-int(len(faction_idx)*0.02):-1]
        hierarchy = hierarchy.at[faction_idx[top_ksiaze]].set(2)

    print(f"[HIERARCHIA] Cykl {cycle}: Nadano rangi – Królów: {jnp.sum(hierarchy == 3)} | Szlachty: {jnp.sum(hierarchy == 1)}")
    return hierarchy

def king_influence(traits, hierarchy, cycle):
    """Król daje boost całej frakcji, ale sam płaci cenę"""
    king_mask = (hierarchy == 3)
    if jnp.any(king_mask):
        # Boost dla całej frakcji króla
        king_faction = factions[king_mask][0]
        faction_mask = (factions == king_faction)
        traits = traits.at[faction_mask, [2, 6, 9]].multiply(1.18)  # + INT, Przebiegłość, Percepcja

        # Koszt dla króla (paranoja, zmęczenie władzą)
        traits = traits.at[king_mask, [3, 8]].multiply(0.82)  # -HP i Reakcje
        print(f"[KRÓL] Król frakcji {king_faction} wzmocnił swoją grupę, ale sam osłabł.")
    return traits

def rebellion(traits, hierarchy, factions, cycle):
    """Możliwość obalenia króla"""
    king_mask = (hierarchy == 3)
    if not jnp.any(king_mask):
        return traits, hierarchy

    # Szansa na rebelię = niska Przebiegłość + niska Percepcja u króla
    king_idx = jnp.where(king_mask)[0][0]
    rebellion_chance = (1.0 - traits[king_idx, 6]) * 0.12 + (1.0 - traits[king_idx, 9]) * 0.08
    if jax.random.uniform(jax.random.PRNGKey(cycle + 666)) < rebellion_chance:
        print(f"!!! [REBELIA] Król został obalony! Nowa koronacja w toku...")
        hierarchy = hierarchy.at[king_idx].set(1)  # spada do Szlachty
        # Najlepszy z frakcji zostaje nowym królem
        faction_mask = (factions == factions[king_idx])
        new_king_local = jnp.argmax(traits[faction_mask, 2] + traits[faction_mask, 6])
        new_king = jnp.where(faction_mask)[0][new_king_local]
        hierarchy = hierarchy.at[new_king].set(3)
    return traits, hierarchy

def evolve_cycle_kingdom(cycles=1):
    global traits, factions, hierarchy, legends
    for c in range(1, cycles + 1):
        print(f"\n=== CYKL KRÓLEWSKI {c} – HIERARCHIA I WŁADZA ===")

        hierarchy = assign_hierarchy(traits, factions, c)
        traits = king_influence(traits, hierarchy, c)
        traits, hierarchy = rebellion(traits, hierarchy, factions, c)

        # Reszta mechanik z poprzednich wersji
        traits = apply_tradeoffs(traits)
        traits = internal_competition(traits)
        traits = apply_mutations(traits)
        traits = level_up(traits)
        legends = create_legends(traits, legends, c)

        # Zapis pełnego stanu
        swarm_data = {
            "traits": traits,
            "factions": factions,
            "legends": legends,
            "hierarchy": hierarchy
        }
        with open(SWARM_STATE_PATH, 'wb') as f:
            pickle.dump(swarm_data, f)

        mean_stats = jnp.mean(traits, axis=0)
        print("ŚREDNIE STATYSTYKI KRÓLESTWA:")
        for name, val in zip(STAT_NAMES, mean_stats):
            print(f"  {name:15} → {val:.3f}")
        print(f"Królów: {jnp.sum(hierarchy == 3)} | Szlachty: {jnp.sum(hierarchy == 1)}")

# === URUCHOM ===
evolve_cycle_kingdom(cycles=4)
```

```python
import jax
import jax.numpy as jnp
import pickle
import os
import time
from datetime import datetime

# === CYWILIZACJA v4 – DYNASTIE, WOJNY, EKONOMIA, EPOKI ===
EPOCHS = ["KAMIENNA", "BRĄZOWA", "MIEDZIANA", "ŻELAZNA", "STALOWA"]
EPOCH_BOOST = [1.0, 1.25, 1.45, 1.70, 2.0]  # progresja technologiczna

# Nowe struktury
swarm_data = load_swarm_with_hierarchy() if 'load_swarm_with_hierarchy' in globals() else {"traits": traits, "factions": factions, "legends": legends, "hierarchy": hierarchy}
traits = swarm_data["traits"]
factions = swarm_data["factions"]
hierarchy = swarm_data["hierarchy"]
legends = swarm_data["legends"]

# Nowe pola cywilizacji
if "economy" not in swarm_data:
    swarm_data["economy"] = jnp.ones(4) * 100.0          # wealth dla 4 frakcji
    swarm_data["reign_years"] = jnp.zeros(N_AGENTS)      # lata panowania króla
    swarm_data["current_epoch"] = 0                       # epoka globalna
    swarm_data["kings"] = jnp.full(4, -1, dtype=int)     # index króla każdej frakcji

economy = swarm_data["economy"]
reign_years = swarm_data["reign_years"]
current_epoch = swarm_data["current_epoch"]
kings = swarm_data["kings"]

KING_NAMES = ["Kragor Wielki", "Sylvara Żelazna", "Thorne Burzowy", "Veyra Mądra", "Zenthos Cień", "Nyxara Płomień"]

def coronation_ceremony(faction, new_king_idx, cycle):
    """Ceremonia koronacji – boost i koszt"""
    global traits
    print(f"⚔️ [KORONACJA] Frakcja {faction} koronuje {KING_NAMES[faction]} (cykl {cycle})")
    # Boost całej frakcji
    faction_mask = (factions == faction)
    traits = traits.at[faction_mask, [2, 6, 9]].multiply(1.22)   # INT, Przebiegłość, Percepcja
    # Koszt dla nowego króla
    traits = traits.at[new_king_idx, [3, 8]].multiply(0.78)      # -HP i Reakcje
    return new_king_idx

def dynasty_inheritance(faction, cycle):
    """Król przekazuje koronę synowi (dziedziczenie)"""
    global kings, hierarchy
    king_idx = kings[faction]
    if king_idx == -1:
        return

    # Szansa na abdykację / śmierć
    if reign_years[king_idx] > 8 and jax.random.uniform(jax.random.PRNGKey(cycle)) < 0.45:
        print(f"👑 [DYNASTIA] Król frakcji {faction} przekazuje koronę synowi!")
        # Znajdź najsilniejszego „syna” (wysoki INT + Przebiegłość)
        faction_mask = (factions == faction)
        candidates = jnp.where(faction_mask & (hierarchy < 3))[0]
        if len(candidates) == 0:
            return
        heir = candidates[jnp.argmax(traits[candidates, 2] + traits[candidates, 6])]

        # Stary król spada do Szlachty
        hierarchy = hierarchy.at[king_idx].set(1)
        # Nowy król
        hierarchy = hierarchy.at[heir].set(3)
        kings = kings.at[faction].set(heir)
        reign_years = reign_years.at[heir].set(0)
        coronation_ceremony(faction, heir, cycle)

def inter_kingdom_war(cycle):
    """Wojny między królestwami – realne straty"""
    global traits, economy
    if jax.random.uniform(jax.random.PRNGKey(cycle + 1000)) > 0.6:
        return

    # Wybierz dwie walczące frakcje
    f1, f2 = jax.random.choice(jax.random.PRNGKey(cycle), 4, (2,), replace=False)
    print(f"🔥 [WOJNA] Frakcja {f1} atakuje frakcję {f2}!")

    # Straty – usuwamy najsłabszych agentów (śmierć)
    mask1 = (factions == f1)
    mask2 = (factions == f2)
    weak1 = jnp.argsort(traits[mask1, 3])[:int(mask1.sum() * 0.12)]  # 12% najsłabszych
    weak2 = jnp.argsort(traits[mask2, 3])[:int(mask2.sum() * 0.15)]

    global_idx1 = jnp.where(mask1)[0][weak1]
    global_idx2 = jnp.where(mask2)[0][weak2]

    traits = traits.at[global_idx1, 3].set(0.05)  # prawie martwi
    traits = traits.at[global_idx2, 3].set(0.05)

    # Straty ekonomiczne
    economy = economy.at[f1].multiply(0.75)
    economy = economy.at[f2].multiply(0.68)
    print(f"   → Straty: ~{len(global_idx1)+len(global_idx2)} agentów, gospodarka w ruinie")

def economy_cycle(cycle):
    """Ekonomia – zarobki, kradzieże, podatki"""
    global economy
    # Bazowe zarobki zależne od epoki i INT frakcji
    for f in range(4):
        faction_mask = (factions == f)
        avg_int = jnp.mean(traits[faction_mask, 2])
        income = 25 * EPOCH_BOOST[current_epoch] * avg_int
        # Kradzieże (Przebiegłość frakcji)
        theft = jnp.mean(traits[faction_mask, 6]) * 12
        economy = economy.at[f].add(income - theft)
        # Podatki od Ludu
        if jnp.sum(faction_mask & (hierarchy == 0)) > 0:
            economy = economy.at[f].add(8)
    print(f"[EKONOMIA] Cykl {cycle} | Bogactwo frakcji: {economy}")

def tech_progress(cycle):
    """Postęp epok – rozwija się z czasem"""
    global current_epoch
    if cycle % 12 == 0 and current_epoch < len(EPOCHS)-1:
        current_epoch += 1
        print(f"🚀 [EPOKA] Przejście do epoki {EPOCHS[current_epoch]}! Technologia skoczyła.")
        # Globalny boost
        traits = traits.at[:, [2, 7]].multiply(1.15)  # INT i Szybkość Nauki

def evolve_cycle_civilization(cycles=1):
    global traits, factions, hierarchy, legends, economy, reign_years, kings, current_epoch
    for c in range(1, cycles + 1):
        print(f"\n=== CYKL CYWILIZACYJNY {c} – DYNASTIE I KRÓLESTWA ===")

        # Podstawowe mechaniki
        hierarchy = assign_hierarchy(traits, factions, c)
        traits = king_influence(traits, hierarchy, c)
        traits, hierarchy = rebellion(traits, hierarchy, factions, c)

        # Nowe systemy
        dynasty_inheritance(0, c)   # frakcja 0
        dynasty_inheritance(1, c)
        dynasty_inheritance(2, c)
        dynasty_inheritance(3, c)

        inter_kingdom_war(c)
        economy_cycle(c)
        tech_progress(c)

        # Zapis pełnego stanu cywilizacji
        swarm_data = {
            "traits": traits, "factions": factions, "legends": legends,
            "hierarchy": hierarchy, "economy": economy,
            "reign_years": reign_years, "kings": kings,
            "current_epoch": current_epoch
        }
        with open(SWARM_STATE_PATH, 'wb') as f:
            pickle.dump(swarm_data, f)

        mean_stats = jnp.mean(traits, axis=0)
        print("STAN CYWILIZACJI:")
        for name, val in zip(STAT_NAMES, mean_stats):
            print(f"  {name:15} → {val:.3f}")
        print(f"Epoka: {EPOCHS[current_epoch]} | Bogactwo ogółem: {jnp.sum(economy):.1f}")

# === URUCHOM ===
evolve_cycle_civilization(cycles=5)
```

```python
import jax
import jax.numpy as jnp
import pickle
import os
import time
from datetime import datetime

# === CYWILIZACJA v5 – DYNASTIE, RELIGIA, PRAWO, HANDEL, PŁEĆ, MAPA ===
# Nowe struktury
swarm_data = load_swarm_with_hierarchy() if 'load_swarm_with_hierarchy' in globals() else {"traits": traits, "factions": factions, "legends": legends, "hierarchy": hierarchy, "economy": economy}
traits = swarm_data["traits"]
factions = swarm_data["factions"]
hierarchy = swarm_data["hierarchy"]
legends = swarm_data["legends"]
economy = swarm_data.get("economy", jnp.ones(4) * 150.0)

# === NOWE POLA ===
if "gender" not in swarm_data:
    swarm_data["gender"] = jax.random.bernoulli(jax.random.PRNGKey(42), p=0.5, shape=(N_AGENTS,))  # 0 = M, 1 = K
    swarm_data["religion"] = jnp.zeros(N_AGENTS, dtype=int)        # 0=Brak, 1=Stara Wiara, 2=Nowy Porządek, 3=Cień
    swarm_data["dynasty"] = jnp.full(N_AGENTS, -1, dtype=int)     # numer dynastii
    swarm_data["house"] = jnp.zeros(N_AGENTS, dtype=int)          # 0=dom, 1=bezdomny
    swarm_data["map_pos"] = jax.random.uniform(jax.random.PRNGKey(123), (N_AGENTS, 2), minval=0, maxval=100)  # prosta mapa 100x100
    swarm_data["season"] = 0                                       # 0=Wiosna, 1=Lato, 2=Jesień, 3=Zima
    swarm_data["day_cycle"] = 0                                    # 0=Dzień, 1=Noc

gender = swarm_data["gender"]
religion = swarm_data["religion"]
dynasty = swarm_data["dynasty"]
house = swarm_data["house"]
map_pos = swarm_data["map_pos"]
season = swarm_data["season"]
day_cycle = swarm_data["day_cycle"]

DYNASTY_NAMES = ["Dom Kragor", "Ród Sylvara", "Klan Thorne", "Rodzina Veyra", "Linia Zenthos"]
HERBS = ["🛡️", "🔥", "🌳", "⚡", "🌑"]  # herby

RELIGION_NAMES = ["Brak wiary", "Stara Wiara", "Nowy Porządek", "Kult Cienia"]

def assign_dynasties_and_herbs(cycle):
    global dynasty
    for f in range(4):
        king_mask = (factions == f) & (hierarchy == 3)
        if jnp.any(king_mask):
            king_idx = jnp.where(king_mask)[0][0]
            dynasty = dynasty.at[king_idx].set(f)
            print(f"🏛️ [DYNASTIA] {DYNASTY_NAMES[f]} ({HERBS[f]}) panuje nad frakcją {f}")

def gender_and_births(cycle):
    global traits, gender, N_AGENTS
    # Narodziny – tylko w frakcjach z domami i kobietami
    female_mask = (gender == 1) & (house == 0)
    birth_chance = (traits[female_mask, 3] * 0.4) * 0.012  # HP i stabilność
    new_borns = jax.random.bernoulli(jax.random.PRNGKey(cycle), p=birth_chance.sum())
    if new_borns > 0:
        # Nowi agenci dziedziczą cechy od matki + ojca (losowy)
        mother_idx = jnp.where(female_mask)[0][jax.random.randint(jax.random.PRNGKey(cycle), (), 0, len(female_mask))]
        father_idx = jnp.where((gender == 0) & (factions == factions[mother_idx]))[0][0]
        child_traits = (traits[mother_idx] + traits[father_idx]) / 2 + jax.random.normal(jax.random.PRNGKey(cycle), (10,)) * 0.1
        child_traits = jnp.clip(child_traits, 0.2, 0.95)
        # Dodajemy dziecko
        traits = jnp.vstack([traits, child_traits])
        gender = jnp.append(gender, jax.random.bernoulli(jax.random.PRNGKey(cycle+1), p=0.5))
        factions = jnp.append(factions, factions[mother_idx])
        print(f"👶 [NARODZINY] Urodziło się {new_borns} dziecko z dziedziczeniem cech!")

def law_and_crime(cycle):
    """Policja, sędziowie, bandyci, dilerzy"""
    # Bandyci i dilerzy – agenci z wysoką Przebiegłością i niską Percepcją
    criminal_mask = (traits[:, 6] > 0.75) & (traits[:, 9] < 0.4)
    if jnp.sum(criminal_mask) > 0:
        print(f"💰 [PRZESTĘPCZOŚĆ] {jnp.sum(criminal_mask)} bandytów/dilerów działa w cieniu")
        economy = economy.at[factions[criminal_mask][0]].add(15)  # zysk z nielegalnego handlu
    # Policja i sędziowie – agenci z wysoką Percepcją
    law_mask = (traits[:, 9] > 0.8) & (hierarchy > 0)
    if jnp.any(law_mask):
        traits = traits.at[law_mask, 6].multiply(1.1)  # + Przebiegłość sędziów

def trade_between_factions(cycle):
    """Legalny i nielegalny handel"""
    for f1 in range(4):
        for f2 in range(f1+1, 4):
            if jax.random.uniform(jax.random.PRNGKey(cycle + f1 + f2)) < 0.7:
                # Legalny handel
                profit = 22 * EPOCH_BOOST[current_epoch]
                economy = economy.at[f1].add(profit)
                economy = economy.at[f2].add(profit)
                print(f"🤝 [HANDEL] Frakcja {f1} handluje z {f2} (+{profit:.0f} bogactwa)")
            elif jax.random.uniform(jax.random.PRNGKey(cycle + 500)) < 0.3:
                # Nielegalny handel / przemyt
                economy = economy.at[f1].add(35)
                print(f"🕵️ [NIELEGALNY HANDEL] Przemyt między {f1} i {f2}")

def seasons_and_day(cycle):
    global season, day_cycle
    season = (cycle // 6) % 4
    day_cycle = cycle % 2
    season_name = ["Wiosna 🌱", "Lato ☀️", "Jesień 🍂", "Zima ❄️"][season]
    print(f"🌍 [PORY ROKU] Aktualnie: {season_name} | {'Dzień' if day_cycle == 0 else 'Noc'}")

def evolve_cycle_full_civilization(cycles=1):
    global traits, factions, hierarchy, legends, economy, dynasty, house, religion
    for c in range(1, cycles + 1):
        print(f"\n=== PEŁNY CYKL CYWILIZACYJNY {c} – DYNASTIE, RELIGIA, PRAWO ===")

        assign_dynasties_and_herbs(c)
        gender_and_births(c)
        law_and_crime(c)
        trade_between_factions(c)
        seasons_and_day(c)

        # Reszta mechanik
        hierarchy = assign_hierarchy(traits, factions, c)
        traits = king_influence(traits, hierarchy, c)
        traits, hierarchy = rebellion(traits, hierarchy, factions, c)
        dynasty_inheritance(0, c)
        inter_kingdom_war(c)
        economy_cycle(c)
        tech_progress(c)

        # Zapis wszystkiego
        swarm_data = {
            "traits": traits, "factions": factions, "legends": legends,
            "hierarchy": hierarchy, "economy": economy, "gender": gender,
            "religion": religion, "dynasty": dynasty, "house": house,
            "map_pos": map_pos, "season": season, "day_cycle": day_cycle,
            "current_epoch": current_epoch
        }
        with open(SWARM_STATE_PATH, 'wb') as f:
            pickle.dump(swarm_data, f)

        print(f"Epoka: {EPOCHS[current_epoch]} | Bogactwo ogółem: {jnp.sum(economy):.0f}")

# === URUCHOM ===
evolve_cycle_full_civilization(cycles=4)
```

```python
import jax
import jax.numpy as jnp
import pickle
import os
import time
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output

# === CYWILIZACJA v5.11 – DYNAMICZNA MAPA, DROGI, HANDEL, KARY, KONTUZJE ===
swarm_data = load_swarm_with_hierarchy() if 'load_swarm_with_hierarchy' in globals() else {"traits": traits, "factions": factions, "hierarchy": hierarchy}
traits = swarm_data["traits"]
factions = swarm_data["factions"]
hierarchy = swarm_data["hierarchy"]
economy = swarm_data.get("economy", jnp.ones(4) * 300.0)

# Nowe pola
if "roads" not in swarm_data:
    swarm_data["roads"] = jnp.zeros((12, 12), dtype=int)          # 0=brak drogi, 1=szlak handlowy, 2=szlak wojskowy
    swarm_data["injuries"] = jnp.zeros(N_AGENTS, dtype=float)     # 0-1.0 poziom kontuzji
    swarm_data["punishments"] = jnp.zeros(N_AGENTS, dtype=int)    # 0=brak, 1=grzywna, 2=banicja, 3=egzekucja
    swarm_data["trade_routes"] = jnp.zeros((4, 4), dtype=float)   # handel między frakcjami

roads = swarm_data["roads"]
injuries = swarm_data["injuries"]
punishments = swarm_data["punishments"]
trade_routes = swarm_data["trade_routes"]

def build_road(f1, f2, cycle):
    """Budowa dróg między frakcjami"""
    global roads
    roads = roads.at[f1, f2].set(1)
    roads = roads.at[f2, f1].set(1)
    print(f"🛤️ [DROGA] Powstał szlak handlowy między frakcją {f1} a {f2}")

def trade_caravan(cycle):
    """Karawany handlowe – legalne i nielegalne"""
    global economy, trade_routes
    for f1 in range(4):
        for f2 in range(f1+1, 4):
            if roads[f1, f2] > 0:  # tylko jeśli jest droga
                profit = 45 * EPOCH_BOOST[current_epoch]
                risk = jax.random.uniform(jax.random.PRNGKey(cycle + f1)) * 0.3
                if risk < 0.15:  # napad
                    economy = economy.at[f1].add(profit * 0.4)
                    economy = economy.at[f2].add(profit * 0.4)
                    print(f"💰 [KARAWANA] Napad na szlak {f1}-{f2}! Straty.")
                else:
                    economy = economy.at[f1].add(profit)
                    economy = economy.at[f2].add(profit)
                    trade_routes = trade_routes.at[f1, f2].add(1)
                    print(f"🤝 [HANDEL] Karawana z frakcji {f1} do {f2} przyniosła zysk.")

def injury_system(cycle):
    """Kontuzje – wpływ na statystyki i gojenie"""
    global injuries, traits
    # Nowe kontuzje w walce lub pracy
    injury_chance = (traits[:, 8] < 0.5) * 0.08  # niska Reakcja = większa szansa
    new_inj = jax.random.bernoulli(jax.random.PRNGKey(cycle), p=injury_chance)
    injuries = injuries.at[new_inj].add(0.35)

    # Gojenie z czasem
    injuries = jnp.clip(injuries - 0.08, 0.0, 1.0)

    # Efekt kontuzji
    traits = traits.at[injuries > 0.6, 3].multiply(0.75)   # -HP
    traits = traits.at[injuries > 0.4, 8].multiply(0.85)   # -Reakcje
    print(f"🩸 [KONTUZJE] {jnp.sum(injuries > 0.3)} agentów rannych (goją się powoli)")

def law_and_punishment(cycle):
    """Sądy, kary, nagrody"""
    global punishments
    criminal_mask = (traits[:, 6] > 0.8) & (injuries < 0.3)  # przebiegli i zdrowi
    if jnp.any(criminal_mask):
        court_idx = jnp.where(criminal_mask)[0][0]
        verdict = jax.random.choice(jax.random.PRNGKey(cycle), 4, p=jnp.array([0.4, 0.3, 0.2, 0.1]))  # grzywna, banicja, egzekucja, uniewinnienie
        punishments = punishments.at[court_idx].set(verdict)
        if verdict == 1:
            economy = economy.at[factions[court_idx]].add(30)  # grzywna
            print(f"⚖️ [SĄD] Grzywna na agenta z frakcji {factions[court_idx]}")
        elif verdict == 2:
            print(f"🚫 [BANICJA] Agent wygnany z frakcji")
            factions = factions.at[court_idx].set(-1)  # bez frakcji
        elif verdict == 3:
            print(f"🪓 [EGZEKUCJA] Agent stracony")
            traits = traits.at[court_idx, 3].set(0.01)  # prawie martwy

def draw_detailed_map(cycle):
    """Szczegółowa mapa z dynamicznymi elementami"""
    clear_output(wait=True)
    fig, ax = plt.subplots(figsize=(10, 10))
    grid = np.zeros((12, 12))
    for i in range(12):
        for j in range(12):
            idx = i * 12 + j
            if idx >= N_AGENTS: continue
            f = int(factions[idx])
            grid[i, j] = f
            # Ikony
            if buildings[idx] > 0:
                ax.text(j, i, BUILDING_NAMES[buildings[idx]][0], ha='center', va='center', fontsize=11, color='white')
            if roads[i//3, j//3] > 0:  # uproszczone drogi
                ax.plot([j-0.4, j+0.4], [i, i], color='brown', linewidth=2)
            if injuries[idx] > 0.4:
                ax.text(j, i, "🩸", ha='center', va='center', fontsize=8)
    ax.imshow(grid, cmap=plt.cm.get_cmap('tab10'), vmin=0, vmax=4)
    ax.set_title(f"MAPA ŚWIATA Ω-16 | Epoka: {EPOCHS[current_epoch]} | Cykl: {cycle} | Drogi aktywne")
    plt.axis('off')
    plt.show()

def evolve_cycle_empire(cycles=1):
    global cycle
    for c in range(1, cycles + 1):
        cycle = c
        print(f"\n=== CYKL IMPERIUM {c} – MAPA I CYWILIZACJA ===")

        trade_caravan(c)
        injury_system(c)
        law_and_punishment(c)
        religious_event(c)

        if c % 2 == 0:
            draw_detailed_map(c)

        # Reszta mechanik
        hierarchy = assign_hierarchy(traits, factions, c)
        traits = king_influence(traits, hierarchy, c)
        economy_cycle(c)
        tech_progress(c)

        # Zapis
        swarm_data = {**swarm_data, "roads": roads, "injuries": injuries, "punishments": punishments, "trade_routes": trade_routes}
        with open(SWARM_STATE_PATH, 'wb') as f:
            pickle.dump(swarm_data, f)

# === URUCHOM ===
evolve_cycle_empire(cycles=4)

print("\n=== MAPA I SYSTEMY GOTOWE ===")
print("Wpisz: draw_detailed_map(999)  → zobacz aktualną mapę")
print("Wpisz: trade_caravan(999)     → wymuś karawanę")
```
